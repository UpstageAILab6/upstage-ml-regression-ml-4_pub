테스트 데이터

```python
df_test = df.query("is_test == 1")
df_train = df.query("is_test == 0")
```

```python
df_test_unique = df_test.drop_duplicates(subset='아파트_unique_key')
```

```python
df_test_unique[df_test_unique['아파트_unique_key'].isnull()]
```

```python
# df_train에서 각 '아파트_unique_key'의 등장 횟수 계산
key_counts = df_train['아파트_unique_key'].value_counts().reset_index()
key_counts.columns = ['아파트_unique_key', 'count']
```

```python
# 등장 횟수를 딕셔너리로 계산
key_counts = df_train['아파트_unique_key'].value_counts().to_dict()

rank = {}
# for문을 사용해 출력
for key in df_test_unique['아파트_unique_key']:
    count = key_counts.get(key, 0)  # 딕셔너리에서 값 가져오기, 없으면 0
    print(f"아파트_unique_key {key}: {count}번 등장")
    rank[key] = count

# rank 딕셔너리를 DataFrame으로 변환
rank_df = pd.DataFrame(list(rank.items()), columns=['아파트_unique_key', 'count'])
```

```python
rank_sorted = rank_df.sort_values(by='count', ascending=True)
```
